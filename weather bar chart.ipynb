```python
import pandas as pd

file_path = r'C:\coupons.csv'
df = pd.read_csv(file_path ,sep=',') 
# Step 2: Summarize the data
# Example: Group by 'Category' and sum the 'Values'
summary = df.groupby('weather')['Y'].sum().reset_index()
import matplotlib.pyplot as plt

# Step 3: Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(summary['weather'], summary['Y'], color='skyblue')
plt.xlabel('Weather')
plt.ylabel('No of Coupons')
plt.title('Bar Chart of Summarized Coupons  by Weather')
plt.xticks(rotation=45)
plt.tight_layout()

# Step 4: Show the plot
plt.show()

import matplotlib.pyplot as plt
%matplotlib inline


plt.show()
```
